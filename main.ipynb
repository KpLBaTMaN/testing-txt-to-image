```python
# Export Directory

export_directory = "/content/drive/My Drive/Generated_Images/"
```

```python
#python program to check if a directory exists
import os


def export_file(prompt_input, images_num):
  print(prompt_input)

  export_file_path = f"{export_directory}{prompt_input}/"


  # Check whether the specified path exists or not
  isExist = os.path.exists(export_file_path)
  if not isExist:

    # Create a new directory because it does not exist
    os.makedirs(export_file_path)
    print("The new directory is created!")

  for x in range(images_num):
    with autocast("cuda"):
      image = pipe(prompt_input)["sample"][0]  
        
    image.save(f"{export_file_path}{prompt_input}_{x}.png")

  
```

```python
array_prompts = [
    # "Rainbow car flying into a black hole",
    # "Thor underwater with rainbow blood",
    "Thanos Walter white"
]

for prompt_instance in array_prompts:
  export_file(prompt_instance,50)
```
